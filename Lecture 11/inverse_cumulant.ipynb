# Generating ranodm numbers with given distribution

## Problem 1

Let $U$ be a random variable uniformly distributed on the interval  $[0, 1)$. Let $Y$ be defined as

$$ Y = \log\left(\frac{u}{1-u}\right)$$

What is the probability distribution of $Y$ ? 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

u = np.linspace(0.0, 1,999)
y = np.log(u/(1-u) )

print("y in ({:},{:})".format(y.min(), y.max()))

plt.plot(u,y)
```

Hint: Use the definition of the cumulant of the probability distribution: 


$$CDF_X(x)=P(X\le x)$$  

and relation:

$$PDF_X(x) = \frac{\text{d}}{\text{d}x} CDF_X(x)$$

```python
n = 10000
bins = int(np.sqrt(n))

u_samples = np.random.uniform(low=0.0, high=1.0, size=n)
y_samples = np.exp(u)/(1+np.exp(u))**2

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10,5), sharex=False)
axs[0].hist(u_samples, bins = bins, color = 'pink')
axs[0].set_title('uniform distribution')
axs[1].hist(y_samples, bins = bins, color = 'purple')
axs[1].set_title('Y = log(u/(1-u)) distribution')
fig.suptitle('Problem 1: Histograms for n = {:} samples\n\n'.format(n))
plt.show()
```
